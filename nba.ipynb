```python
# Fill missing values for Salary with the median
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
```

```python
print(df['Salary'])
```

```python
# Fill missing values for College with 'Unknown'

df['College'] = df["College"].fillna("unknown")
print(df['College'])
```

```python
# Drop rows where critical information (like Name or Team) is missing

df = df.dropna(subset=['Name', 'Team'])
print(df.isnull().sum())
print(df.info())
```

```python
# Convert 'Height' to numerical format
def convertHeightStringToNum(Heightstr):
    if pd.isna(Heightstr):
        return np.nan
    feet, inches = Heightstr.split('-')
    return int(feet) * 12 + int(inches)
```

```python
feet, inches = 12, 8

print("hello-hy".split('-'))
```

```python
df['Height'] = df['Height'].apply(convertHeightStringToNum)
print(df['Height'])
```

```python
print(df['Height'])
```

```python
# Convert 'Number' to integer type

df['Number'] = df['Number'].astype(int)
print(df['Number'])
```

```python
# Convert 'Salary' to integer type
df['Salary'] = df['Salary'].astype(int)
```

```python
# Convert 'Age' to integer type
df['Age'] = df['Age'].astype(int)
print(df["Age"])
```

```python
# Standardize team names
df['Team'] = df["Team"].str.strip().str.title()
print(df['Team'])
```

```python
# Standardize position names
print(df['Position'])
df['Position'] = df['Position'].str.strip().str.upper()
print(df['Position'])
```

```python
# Filter players older than 30
olderThan30 = df[df['Age']>30]

# print(olderThan30.value_counts().sum())
olderThan30
```

```python
# Sort players by Salary in descending order
topEarners = df.sort_values(ascending=False, by='Salary')
topEarners
```

```python
# Calculate average salary by team

avgSalaryByteam = df.groupby('Team')["Salary"].mean()

avgSalaryByteam
```

```python

# Set up the plotting style

sns.set_style('whitegrid')
```

```python
# Histogram of player ages

plt.figure(figsize=(10, 10))
sns.histplot(df["Age"], bins=5, kde=True)
plt.title("age distributon")
plt.xlabel("Age")
plt.ylabel("frequency")
plt.show()
```

```python
# Box plot of salaries by team

plt.figure(figsize=(14, 8))
sns.boxplot(x='Team', y="Salary", data=df)
plt.xticks(rotation=90)
plt.title("Salary Distribution By Team")
plt.xlabel("Team")
plt.ylabel("Salary")
plt.show()
```

```python
# Scatter plot of Height vs. Weight
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Height', y='Weight', data=df, hue='Position', palette='viridis')
plt.title('Height vs. Weight of Players')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (lbs)')
plt.legend(loc='upper right')
plt.show()
```
